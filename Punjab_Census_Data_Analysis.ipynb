```python
#Predict Worker Participation Rate
#Step 1: Select Features
features = ['District-wise population and their ranking Population - 2011', 'District - wise sex ratio Census 2011', 'growth_rate']
X = data[features]

# Handle infinite values in 'growth_rate' by replacing them with NaN and then filling with the mean
X.loc[:, 'growth_rate'] = X['growth_rate'].replace([float('inf'), -float('inf')], float('nan'))
X.loc[:, 'growth_rate'] = X['growth_rate'].fillna(X['growth_rate'].mean())

y = data['worker_rate']
```

```python
#Step 2: Train-Test Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
#Step 3: Train Model
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#step 4: Prediction
y_pred = model.predict(X_test)
```

```python
#evaluation
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import r2_score
print("MSE:", mean_squared_error(y_test, y_pred))
print("R2 Score:", r2_score(y_test, y_pred))
```

```python
#PART 2: Clustering
#Select Data

X = data[['District-wise population and their ranking Population - 2011', 'District - wise sex ratio Census 2011', 'worker_rate']]
```

```python
#Apply KMeans
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3)
data['cluster'] = kmeans.fit_predict(X)
```
